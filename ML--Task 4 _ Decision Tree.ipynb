```python
x = iris.iloc[:,0:4]
y = iris.values[:,-1]
```

# Label Encoding :convertion of variable data to numerical data

```python
from sklearn import preprocessing
pred = preprocessing.LabelEncoder()
y = pred.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(x,y)
```

# Performing Decision tree algorithm

```python

from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train , y_train)
```

```python
y_prediction = dt.predict(x_test)
pred1 = pd.DataFrame({'Predicted': y_prediction , 'Actual' : y_test})
pred1
```

# Accuracy check

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_prediction)
print("Accuracy = " +str(acc))
```

# Visualization of DT

```python
from sklearn.tree import plot_tree
fig = plt.figure(figsize = (16,16))
plot_tree(dt, feature_names= x_train.columns,class_names = ['setosa' ,'versicolor','virginca' ] , filled = True , rounded= True)
```
